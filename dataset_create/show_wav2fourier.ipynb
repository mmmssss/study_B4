```python
import os
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np

#Define random seed
np.random.seed(seed=32)
```

```python
def wav_show(num1, num2, num3, num4):

    sn_rate = "sn-4.0"
    fig = plt.figure(figsize=(24, 4))
    plt.subplots_adjust(wspace=0.35, hspace=0.2)
    plt.rcParams["font.size"] = 28

    class_path = "glassbreak"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num1] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 1)
    ax.set_xticks([0, 220500, 441000])
    ax.set_xticklabels(['0', '5', '10'], fontsize='small')
    plt.title("Glass breaking")
    plt.xlabel("time[sec]")
    plt.ylim([-1.1, 1.1])
    plt.yticks([-1.0,0.0,1.0])
    plt.plot(audio)

    class_path = "gunshot"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num2] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 2)
    ax.set_xticks([0, 220500, 441000])
    ax.set_xticklabels(['0', '5', '10'], fontsize='small')
    plt.title("Gunshot")
    plt.xlabel("time[sec]")
    plt.ylim([-1.1, 1.1])
    plt.yticks([-1.0,0.0,1.0])
    plt.plot(audio)

    class_path = "babycry"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num3] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 3)
    ax.set_xticks([0, 220500, 441000])
    ax.set_xticklabels(['0', '5', '10'], fontsize='small')
    plt.title("Baby crying")
    plt.xlabel("time[sec]")
    plt.ylim([-1.1, 1.1])
    plt.yticks([-1.0,0.0,1.0])
    plt.plot(audio)

    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num4] + "/bgm.wav")
    ax = fig.add_subplot(1, 4, 4)
    ax.set_xticks([0, 220500, 441000])
    ax.set_xticklabels(['0', '5', '10'], fontsize='small')
    plt.title("Background audio")
    plt.xlabel("time[sec]")
    plt.ylim([-1.1, 1.1])
    plt.yticks([-1.0,0.0,1.0])
    plt.plot(audio)

    #plt.savefig("ex_dataset.eps")
```

```python
wav_show(11, 39, 12, 12)
```

```python
def fourier_show(data, Fs):
    F = np.fft.fft(data)
    amp = np.abs(F)

    # 周波数軸のデータ作成
    freq = np.linspace(0, Fs, Fs)

    # print(freq.shape)  # --> (16000,)

    plt.xlabel('frequency[Hz]')
    plt.plot(freq[:int(Fs/2)+1], amp[:int(Fs/2)+1]) # ナイキスト定数まで表示
    plt.grid()
    plt.tight_layout()
```

```python
def fourier_shows(num1, num2, num3, num4):
    
    fig = plt.figure(figsize=(28, 6))
    plt.subplots_adjust(wspace=0.35, hspace=0.2)
    plt.rcParams["font.size"] = 28
    sn_rate = "sn-4.0"

    class_path = "glassbreak"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num1] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 1)
    fourier_show(audio, Fs)
    plt.ylabel('amplitude')

    class_path = "gunshot"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num2] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 2)
    fourier_show(audio, Fs)

    class_path = "babycry"
    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num3] + "/truth.wav")
    ax = fig.add_subplot(1, 4, 3)
    fourier_show(audio, Fs)

    path = "./mix_dataset/" + sn_rate + "/" + class_path + "/"
    files = os.listdir(path)
    audio, Fs = sf.read(path + files[num4] + "/bgm.wav")
    ax = fig.add_subplot(1, 4, 4)
    fourier_show(audio, Fs)

    #plt.savefig("ex_dataset.eps")
```
